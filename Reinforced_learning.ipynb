```python
#Reinforcement learning?
#computer learns how to make decisions by interacting with the environment.
''''
The agent/ machine receives rewards for good decision.

Key Components of RL
-> Environment -> Actions -> Agent.

#Real-world Application.
-Self driving cars
-Robots
-Game Playing
-Recomendation systems
-smart traffic lights
-Ware house automation

'''
```

```python
import random

# Environment
goal = 5
position = 0

# Output
print('Robot starts at position:', position)

while position != goal:
    action = random.choice(['LEFT', 'RIGHT'])

    # Action
    if action == 'RIGHT':
        position += 1
    else:
        position = max(0, position - 1)

    # Rewards
    if position == goal:
        reward = 100
    else:
        reward = -1

    # Output
    print(f"Action: {action:5} Position: {position} Reward: {reward}")

print("\nRobot has reached the goal")
```
