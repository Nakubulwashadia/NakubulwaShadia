```python
import pandas as pd
import numpy as np                   #scikit-learn science kit learning for machine learning
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#create sample data for Demonstration
#Feature: size
#Regression for predicting House prices based on square feet
#Size, bedroom, age

data = {
    'Size': [1500, 1800, 2400, 3000, 3500],
    'Bedroom': [3, 4, 3, 5, 4],
    'Age': [10, 15, 20, 5, 8],
    'Price': [400000, 500000, 600000, 650000, 700000]
}

df = pd.DataFrame(data)
```

```python
#prepare data
x = df[['Size', 'Bedroom', 'Age']]  # Features
y = df['Price']                     # Target variable
```

```python
#Split Data
x_train, x_test, y_train, y_train = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
#select and train the model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#Make predictions
y_pred = model.predict(x_test)
```

```python
#Evaluate model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
#Output
print('----Regression Results:')
print(f'Prediction: {y_pred}')
print(f'Actual Prices: {y_test.values}')
print(f'Mean Squared Error: {mse: .2f}')
print(f'R-squared Score: {r2: .2f}') #Is higher than Max 1.0 the better
```
